```python
import json
import pandas as pd
from plotnine import *
```

```python
results_simple = []
with open('sweep_simple.txt') as f:
    for line in f.readlines():
        results_simple.append(json.loads(line))
results_simple = [x for x in results_simple if x['recovery_rate'] > 0.9]
results_simple = sorted(results_simple, key=lambda x: x['throughput'])[::-1]
```

```python
results_qr = []
with open('sweep_qr.txt') as f:
    for line in f.readlines():
        results_qr.append(json.loads(line))
results_qr = [x for x in results_qr if x['recovery_rate'] > 0.9]
results_qr = sorted(results_qr, key=lambda x: x['throughput'])[::-1]
```

```python
files = [
    'alpha0',
    'alpha2',
    'alpha4',
    'alpha6',
    'alpha8',
    'baseline_wifi'
]
data = {'config': [] , 'size': []}
for file in files:
    file_dir = 'skype_packet_recordings/measurement_2/{}.txt'.format(file)
    for line in open(file_dir).readlines():
        try:
            size = int(line.split('Len=')[1])
            data['config'].append(file)
            data['size'].append(size)
        except:
            pass
data = pd.DataFrame(data)
```

```python
ggplot(data) + geom_density(aes(x='size', color='config'))
```
